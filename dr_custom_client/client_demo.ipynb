```python
import pandas as pd
import requests
import json
import time 

EXTERNAL_API = "http://127.0.0.1:5000"
```

```python
# Get a specific DR signal from April 30 to May 1, 2019

DR_SIGNAL = 'dr_prices'
start_date_dr_sig = '2019-04-30T00:00:00'
end_date_dr_sig = '2019-05-01T00:00:00'

df_list = []

response = requests.get(
    '{0}/get-dr-signal/{1}'.format(EXTERNAL_API, DR_SIGNAL),
    params={'startdate': start_date_dr_sig, 'enddate': end_date_dr_sig},
)

for data_sig in json.loads(response.content):
    df = pd.DataFrame(json.loads(data_sig))
    df_list.append(df)

# Show the first one
df_list[0]
```

```python
# All the DR signals for April 30, 2019
start_date_dr_sig = '2019-04-30T00:00:00'
end_date_dr_sig = '2019-04-30T23:59:59'

response = requests.get(
    '{0}/get-all-signal'.format(EXTERNAL_API)
)

df_dict = {}

for data_sig in json.loads(response.content):
    data_json = data_sig
    if data_json['type'] not in df_dict.keys():
        df_dict[data_json['type']] = []
        
    df_dict[data_json['type']].append(pd.DataFrame(json.loads(data_json['data'])))
```
